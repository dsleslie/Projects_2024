```python
# 가맹점_광역시도 컬럼 제거하여 새로운 데이터프레임 생성
cc_23_60_mod = cc_23_60.drop(columns=['거래년월', '가맹점_광역시도', '가맹점_시군구', '승인건수'])

# 결과 출력 (새로운 데이터프레임의 첫 몇 줄을 확인)
print(cc_23_60_mod.head(10))
```

```python
# 고객 ID와 가맹점업종명으로 그룹화하여 승인금액 합산
ID_업종별합산_23_40 = cc_23_40_mod.groupby(['고객ID', '가맹점업종명'])['승인금액'].sum().reset_index()

# 결과 출력 (합산된 데이터프레임의 첫 몇 줄을 확인)
print(ID_업종별합산_23_40)
```

```python
# 가맹점업종명으로 그룹화하여 승인금액 평균 계산
업종별_승인금액_평균_23_40 = ID_업종별합산_23_40.groupby('가맹점업종명')['승인금액'].mean().reset_index()

# 과학적 표기법 활성화 (기본 설정으로 되돌리기)
#pd.options.display.float_format = None

# 과학적 표기법 해제
pd.options.display.float_format = '{:,.0f}'.format

# 승인금액을 기준으로 내림차순 정렬
업종별_승인금액_평균_23_40 = 업종별_승인금액_평균_23_40.sort_values(by='승인금액', ascending=False)

# 결과 출력 (평균 승인금액 데이터프레임의 첫 몇 줄을 확인)
업종별_승인금액_평균_23_40
```
